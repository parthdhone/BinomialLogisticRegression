# **Binomial logistic Regression**

```python
import numpy as np

def sigmoid(x):
  result = 1/ (1 + np.exp(-x))
  return result

sigmoid(-5) # --- Give the output in between 0 - 1
```

#**Multinomial logistic regression** (more than two classes)
* Instead of sigmoid it use softmax
* It converts more than outputs into probablities -- summing up will give you 1
* The class having highest probability will be choosen

**What is softmax function**
it converts the vectors of number(Logits) into probabilities.

image.png

```python
fruits = np.array([
    [150,7], # apple -- 150grams, 7 is sweetness level
    [120,10], # Banana
    [180, 5]  # orange
])
```

Weights:
It represents the importance of each features(mass and sweetness) for predicting the fruit


```python
weight = np.array([
    [0.02,0.04],  # apple 0.02 weight for apple mass and 0.04 for apples sweetness
    [0.03,0.05], # banana
    [0.01,0.02] # Orange
])
```

```python
newFruit = np.array([130,8])
```

```python
def softmax(x):
  expScore = np.exp(x - np.max(x)) # Preventation of overflow
  result = expScore / np.sum(expScore)
  return result
```

# Why we will do the dot product
* It will mutiply each fruits class's weight with the new fruit features
* A raw score (logits) before applying softmax()

```python
score = np.dot(weight,newFruit)
print(score)
```

**Probability**

```python
prob = softmax(score)
prob
```

```python
fruit_class = ['Apple','Banana','Orange']
predictedClass = fruit_class[np.argmax(prob)] # argmax --- returns the index of highest value
predictedClass
```

# S curve

```python
import pandas as pd
x = np.linspace(-10,10,100) # We will do always
y_sigmoid = sigmoid(x)
y_sigmoid
df = pd.DataFrame()
df['x'] = x
df['y_sigmoid'] = y_sigmoid
df
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (7,5))
plt.plot(df['x'], df['y_sigmoid'],color='red')
plt.axhline(y = 0.5 , color = 'Blue', linestyle = '--') # Adds a horizontal line
```
